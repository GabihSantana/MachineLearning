## Exercicio Aula 6:

1. Histograma da coluna total_bill
2. Boxplot da coluna tips
3. Boxplot usando as variaveis total_bill e smoker
4. Violinplot usando as variaveis tips e day

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = sns.load_dataset('tips')
```

```python
dados.head()
```

```python
sns.histplot(dados['total_bill'], kde=True, color='red', stat='count', element='bars', bins=10, shrink=.8)
plt.title('Histograma: total_bill')
```

```python
sns.boxplot(dados['tip'], color='#8ebf49', linecolor='darkgreen', linewidth=2)
plt.title('Boxplot: tips')
```

```python
sns.boxplot(x = 'smoker', y = 'total_bill', data=dados, color='purple')
plt.title('Boxplot: total_bill e smoker')
```

```python
sns.violinplot(x='day', y='tip', data=dados, color='#69d4db', inner='box')
plt.title('Violinplot: tips e day')
```

```python
sns.violinplot(x='day', y='total_bill', data=dados, color='#69A4D0', inner='quart')
plt.title('Violinplot: total_bill e day')
```

```python
sns.kdeplot(dados['total_bill'])
```

```python
sns.rugplot(dados['total_bill'])
```
