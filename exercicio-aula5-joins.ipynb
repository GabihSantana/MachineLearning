## Exercicio - Aula 5 - Joins

```python
import pandas as pd
import numpy as np
```

```python
data1 = {'Nome': ['Anderson', 'Beatriz', 'Bruno', 'Cristiano', 'Daniel', 'Douglas', 'Elana', 'Fabricio'],
         'Chave':['C0','C1','C2','C3','C4','C5','C6','C7'],
         'Chave1':['C0','C1','C2','C3','C0','C1','C2','C3'],
         'Idade': [24,36,45,56,29,20,38,47],
         'Cidade': ['Santos','Atibaia','Guarulhos','Atibaia','Santos','Guarulhos','Guarulhos','Atibaia'],
         'Registro': [32123,45828,34985,13457,98348,99348,45019,24894],
         'Departamento':['RH','PD','RH','ADM','ADM','RH','ADM','PD'],
         'Formacao':['Graduacao','Mestrado','Graduacao','Tecnico','Graduacao','Tecnico','Graduacao','Graduacao']
        }
```

```python
data2 = {'Nome': ['Gabriel', 'Janaina', 'Luiz', 'Marcos', 'Nadia', 'Oswaldo', 'Patricia', 'Ricardo'],
         'Chave':['C0','C1','C2','C3','C4','C5','C6','C7'],
         'Chave1':['C0','C0','C0','C0','C1','C1','C1','C1'],
         'Idade': [23,56,32,22,36,68,22,38],
         'Cidade': ['Guarulhos','Guarulhos','Atibaia','Guarulhos','Santos','Guarulhos','Guarulhos','Santos'],
         'Registro': [32948,13480,22884,56821,65938,12956,75439,45392],
         'Departamento':['PD','ADM','PD','ADM','PD','RH','ADM','PD'],
         'Formacao':['Tecnico','Graduacao','Graduacao','Graduacao','Graduacao','Mestrado','Graduacao','Mestrado']
        }
```

```python
dafra1 = pd.DataFrame(data1)
print(dafra1)
```

```python
dafra2 = pd.DataFrame(data2)
print(dafra2)
```

```python
# Juncao Interna > Inner Join
pd.merge(dafra1, dafra2, how='inner', on=['Chave', 'Chave1'])
```

```python
# Juncao externa a esquerda > Left Join
pd.merge(dafra1, dafra2, how='left', on=['Chave', 'Chave1'])
```

```python
# Juncao externa a direita > Right Join
pd.merge(dafra1,dafra2, how='right', on=['Chave', 'Chave1'])
```

```python
# Juncao externa completa > Outer Join
pd.merge(dafra1,dafra2, how='outer',on=['Chave', 'Chave1'])
```

```python
# Produto Cartesiano > Cross Join
pd.merge(dafra1,dafra2, how='cross')
```
