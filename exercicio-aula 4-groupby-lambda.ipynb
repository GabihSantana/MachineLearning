# Exercicio 1 - Groupby Lambda

```python
import pandas as pd
import numpy as np
```

```python
# Dados em dicionario

dadosDict = {'Nome': ['Anderson', 'Beatriz', 'Bruno', 'Cristiano', 'Daniel', 'Douglas', 'Elena', 'Fabricio', 'Gabriel', 'Janaina', 'Luiz', 'Marcos', 'Nadia', 'Oswaldo', 'Patricia', 'Ricardo'],
        'Idade': [24,36,45,56,29,20,38,47,23,56,32,22,36,68,22,38],
        'Cidade': ['Santos', 'Atibaia', 'Guarulhos', 'Atibaia', 'Santos', 'Guarulhos','Guarulhos', 'Atibaia','Guarulhos','Guarulhos', 'Atibaia', 'Guaurlhos', 'Santos', 'Guarulhos', 'Guarulhos', 'Santos'],
        'Registro': [32123, 45828,34985,13457,98439,99348,45019,24894,32948,13480,22884,56821,65938,12956,75439,45392],
        'Departamento': ['RH','PD','RH','ADM','ADM','RH','ADM','PD','PD','ADM','PD','ADM','PD','RH','ADM','PD'],
        'Formacao': ['Graduacao', 'Mestrado', 'Graduacao', 'Tecnico','Graduacao', 'Tecnico','Graduacao','Graduacao','Tecnico', 'Graduacao','Graduacao','Graduacao','Graduacao', 'Mestrado','Graduacao','Mestrado']
        }
```

```python
# Gerando o DataFrame

dafra = pd.DataFrame(dadosDict)
print(dadosDict)
```

```python
# Agrupamento de Cidades
cidadeFuncionarios = dafra.groupby(['Cidade'])
print(cidadeFuncionarios.groups)
```

```python
# Agrupamento de Formacao
formacaoFuncionarios = dafra.groupby(['Formacao'])
print(formacaoFuncionarios.groups)
```

```python
# Agrupamento de Departamento
departamentoFuncionarios = dafra.groupby(['Departamento'])
print(departamentoFuncionarios.groups)
```

```python
# Agrupamento de Cidade e Departamento
cidadeEDepartamento = dafra.groupby(['Cidade','Departamento'])
print(cidadeEDepartamento.groups)
```

```python
# Agrupamento de Departamento e Formacao
departamentoEGrupos = dafra.groupby(['Departamento','Formacao'])
print(departamentoEGrupos.groups)
```

```python
# Calculo da diferenca entre a media e o valor de cada subgrupo
mediaLambda = lambda x: (x.mean()-x)
```

```python
cidadeFuncionarios.transform(mediaLambda)
```

```python
formacaoFuncionarios.transform(mediaLambda)
```

```python
departamentoFuncionarios.transform(mediaLambda)
```

```python
cidadeEDepartamento.transform(mediaLambda)
```

```python
departamentoEGrupos.transform(mediaLambda)
```

```python
departamentoFuncionarios.mean()
```

```python
# Problemas com kernel
cidadeFuncionarios_t = dafra.groupby('Cidade')[['Idade', 'Registro']]
cidadeFuncionarios_t.transform(mediaLambda)
```
